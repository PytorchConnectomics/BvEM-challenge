## Introduction
This is a notebook that shows how to evaluate 3D Blood Vessel Segmentation result. We only evaluate the largest instance since the length of the largest instance is around 99%, 95%, and 85% for each dataset.

```python
import numpy as np

from metric import multi_instance_metric, single_instance_metric
```

```python
# random a ground truth and a prediction
gt = np.zeros((256, 256, 256))
gt[100:150, 100:150, 100:150] = 1

pred = np.zeros((256, 256, 256))
pred[90:140, 90:140, 90:140] = 1

# if the prediction is a single instance, use single_instance_metric
precision, recall, acc = single_instance_metric(pred, gt)
print(f'precision: {precision}, recall: {recall}, acc: {acc}')
```

```python
# random a ground truth and a prediction
gt = np.zeros((256, 256, 256))
gt[100:150, 100:150, 100:150] = 1

pred = np.zeros((256, 256, 256))
pred[90:140, 90:140, 90:140] = 1 # instance 1
pred[10:70, 10:70, 10:70] = 1 # instrance 2

# if the prediction is a multi-instance, use multi_instance_metric
precision, recall, acc = multi_instance_metric(pred, gt)
print(f'precision: {precision}, recall: {recall}, acc: {acc}')
```
